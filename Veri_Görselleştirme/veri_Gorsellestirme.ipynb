* Bu veriseti NASA'nın yayınladığı, galaksi keşfi ile ilgili bir verisetidir.

* method sütunu galaksilerin bulunma methodunu ifade eder.

* number sütunu bulunan sistemlerdeki gezegen sayısını ifade eder.

* orbital_period sütunu yörünge dönemini ifade eder.

* mass sütunu kütleyi ifade eder.

* distance sütunu uzklığı ifade eder.

* year sütunu bulunma tarihini ifade eder.

# Çizgi Grafik

```python
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

## Veri Setinin Hikayesi

Beyine bağlanan bir cihaz aracılığı ile toplanan sinyalleri ifade eden bir verisetidir.
* **subject:** Verilerin toplandığı kişiyi ifade ediyor.
* **timepoint:** Zaman noktalarını ifade ediyor.
* **event:** Birbirinden farklı olayları ifade ediyor.
* **region:** Sinyalin toplandığı bölgeyi ifade ediyor.
* **signal:** Gelen sinyali ifade ediyor.

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```
